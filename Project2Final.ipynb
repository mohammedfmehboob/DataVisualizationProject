## Purpose/Objectives: 
The question of whether males and females receive equal pay, including the factors of education and age levels, was answered. Data was given from wageGenderEduAge.csv file and the data was divided into different categories. The categories were: wage, education, age, and gender. Graphs were made to see if there were patterns in the data and from those patterns, the question was answered. 

## Getting the Data

```python
# loading libaries that might be needed. 
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading in data
the_data = pd.read_csv("datasets/wageGenderEduAge.csv")
```

```python
# Looking at the data
display(the_data)
```

## Graphs/Results: 

On a general level disregarding education and ages, do both genders get paid the same? This question was answered through a histogram that has a Kinetic Density Plot (KDE) plotted over it. The graph is shown here: 

```python
sns.histplot(the_data, x = 'wage', hue = 'gender',
             kde = True, fill = True, 
             element = 'step', palette = 'tab10', 
             alpha=0.5)
plt.title("Wages by Gender")
plt.xlabel("Wages")
```

**Figure 1.** Histogram of Wages by gender.

The figure above plots Wages on the X-axis and Count on the Y-axis. This figure shows that males in fact do get paid more than females overall. Towards the end where the wages are higher, males have a much higher count than females while the opposite occurs in the lower wages. This is a clear indication that on a general level, males do get paid more than women. 

However, what about education or age? Do both genders get paid equally at those levels? To answer this question, two different scatterplots were made. These two plots can be seen here: 

```python
sns.lmplot(data = the_data, x = "education", 
           y = "wage", hue = "gender",
          scatter_kws = {"s": 5}, palette = "tab10")
plt.xlabel("Education")
plt.ylabel("Wages")

plt.title("Wages vs. Education by Gender")
```

**Figure 2.** Scatterplot of Wages by Education and gender. 

Figure 2 above plots Wages on the Y-axis and Education on the X-axis. From this graph, it can be seen that females are paid less than males overall because the male regresson line is higer than the female line. However, as females gain more education, the wage gap closes. This is clear as the gap between the two regression lines get smaller and the female regression line is a lot steeper than the male line. Despite the wage gap closing, males do get paid more than females at every education level. 

```python
sns.lmplot(data = the_data, x = "age", y = "wage", hue = "gender", 
          scatter_kws = {"s": 5}, palette = 'tab10')
plt.xlabel("Age")
plt.ylabel('Wages')
plt.title("Wages vs. Age by Gender")
```

**Figure 3.** Scatterplot of Wages by Age and gender.

Figure 3 plots Wages on the y-axis and Ages on the X-axis. This figure reinforces the idea that males get paid more than females, regardless of age. On both graphs, the male regression line is a lot higher than the femaleâ€™s line throughout the entire x-axis. Furthermore, the slopes of the regression line of the males is steeper than the females, indicating more growth compared to females. Due to this,the wage gap actually increases between males and females as age increases, showing a bias for males. Due to this, it can be confirmed that males do get paid more than females. 

However, is that the only evidence there is? Well, joint plots with KDEs were made to further indicate if this conclusion is true: 

```python
g =sns.jointplot(x = 'age', y = 'wage', data = the_data, 
                 kind = 'kde', fill = True, hue = 'gender', 
                 alpha = 0.8, 
                palette = 'tab10')
g.ax_joint.set_xlabel('Age')
g.ax_joint.set_ylabel('Wages')
g.fig.suptitle("Joint Plot of Wages by Age and Gender")
g.fig.subplots_adjust(top=0.95) # Reduce plot to make room
```

**Figure 4.** Joint plot of Wages by Age and Gender

```python
g = sns.jointplot(x = 'education', y = 'wage', data = the_data, kind ='kde', fill = True, hue = 'gender', alpha = 0.8, 
              palette = 'tab10')
g.ax_joint.set_xlabel('Education')
g.ax_joint.set_ylabel('Wages')
g.fig.suptitle("Joint Plot of Wages by Education and Gender")
g.fig.subplots_adjust(top=0.95) # Reduce plot to make room
```

**Figure 5.** Joint plot of Wages by Education and Gender.

Figure 4 plotted Wages on the Y-axis and Age on the X-axis. Figure 5 plotted Wages on the Y-axis and Education on the X-axis. Both of these figures provide further evidence for the idea that there is a wage gap between males and females. The males dominate both factors (education and age) because the graphs are dominantly orange (male) rather than blue (female). This indicates that there is a higher count for males than females at most of the wages. There are a few hints of blue areas in the higher wages (see Figure 5), but those areas are also prominently orange. Even though education does yield better wages for females, it does not equate to the number of males in those regions and males have better wages at the same education levels. So, these joint plots support the ideas that were stated before.

## Conclusion

All the evidence and patterns show that males get paid more than females in general, regardless of education or age. The males have higher growth in wages compared to females as their age increases. However, females do close the wage gap as they receive more education, but males get paid more at each education level regardless. Overall, it can be seen that males get paid more than females and that there is a wage gap between them.
