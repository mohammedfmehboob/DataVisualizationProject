## Purpose/Objectives: 
The question of whether males and females receive equal pay, including the factors of education and age levels, was answered. Data was given from wageGenderEduAge.csv file and the data were divided into different categories. The categories were: wage, education, age, and gender. Graphs were made to see if there were patterns in the data and from those patterns, the question was answered. 

## Getting the Data

```python
# loading libaries that might be needed. 
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading in data
the_data = pd.read_csv("datasets/wageGenderEduAge.csv")
```

```python
# Looking at the data
display(the_data)
```

## Graphs/Results: 

On a general level disregarding education and ages, do both genders get paid the same? This question was answered through a histogram that has a Kinetic Density Plot (KDE) plotted over it. The graph is shown here: 

```python
sns.histplot(the_data, x = 'wage', hue = 'gender',
             kde = True, fill = True, 
             element = 'step', palette = 'tab10', 
             alpha=0.5)
plt.title("Wages by Gender")
plt.xlabel("Wages")
```

**Figure 1.** Histogram of wages by gender.

The figure above shows that males in fact do get paid more than females overall. Towards the end where the wages are higher, males have a much higher count than females while the opposite occurs in the lower wages. This is a clear indication that on a general level, males do get paid more than women. 

However, what about education or age? Do both genders get paid equally at those levels? To answer this question, two different scatterplots were made. These two plots can be seen here: 

```python
sns.lmplot(data = the_data, x = "education", 
           y = "wage", hue = "gender",
          scatter_kws = {"s": 5}, palette = 'tab10')
plt.xlabel("Education")
plt.ylabel("Wages")
plt.title("Wages vs. Education by Gender")
```

**Figure 2.** Scatterplot of wages by education and gender. 

```python
sns.lmplot(data = the_data, x = "age", y = "wage", hue = "gender", 
          scatter_kws = {"s": 5}, palette = 'tab10')
plt.xlabel("Age")
plt.ylabel('Wages')
plt.title("Wages vs. Age by Gender")
```

**Figure 3.** Scatterplot of wages by age and gender.

Figures 2 and 3 above reinforce the idea that males get paid more than females, regardless of education or age. On both graphs, the male regression line is a lot higher than the femaleâ€™s line throughout the entire x-axis. Furthermore, the slopes of the regression lines of the males is steeper than the females, indicating more growth compared to females. Due to this, it can be confirmed that males do get paid more than females. 

However, is that the only evidence there is? Well, joint plots were made to further indicate if this conclusion is true: 

```python
g =sns.jointplot(x = 'age', y = 'wage', data = the_data, 
                 kind = 'kde', fill = True, hue = 'gender', 
                 alpha = 0.8, 
                palette = 'tab10')
g.ax_joint.set_xlabel('Age')
g.ax_joint.set_ylabel('Wages')
g.fig.suptitle("Joint Plot of Wages by Age and Gender")
g.fig.subplots_adjust(top=0.95) # Reduce plot to make room
```

**Figure 4.** Joint plot of Wages by Age and Gender

```python
g = sns.jointplot(x = 'education', y = 'wage', data = the_data, kind ='kde', fill = True, hue = 'gender', alpha = 0.8, 
              palette = 'tab10')
g.ax_joint.set_xlabel('Education')
g.ax_joint.set_ylabel('Wages')
g.fig.suptitle("Joint Plot of Wages by Education and Gender")
g.fig.subplots_adjust(top=0.95) # Reduce plot to make room
```

**Figure 5.** Joint plot of Wages by Education and Gender.

Figures 4 and 5 above demonstrate the same conclusion as before. The males dominate both factors because the graphs are dominantly orange (male) rather than blue (female). Furthermore, towards the higher wages, those areas are also prominently orange with a few hints of blue. Even though education does have slightly better results for females, the overall general conclusion is that males do in fact get paid more than women.

## Conclusion

All the evidence and patterns show that males get paid more than females in general, regardless of education or age. The males have higher growth in wages compared to females as the number of years and age goes up for the participants. Overall, males get paid more than females.
