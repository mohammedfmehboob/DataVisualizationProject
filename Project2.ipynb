```python
# loading libaries that might be needed. 
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
the_data = pd.read_csv("datasets/wageGenderEduAge.csv")
```

```python
# Looking at the data
display(the_data)
```

```python
# using a violin plot to see if there is a general difference.
sns.violinplot(data = the_data, x = 'gender', y = 'wage')
```

```python
sns.histplot(the_data, x = 'wage', hue = 'gender', kde = True, fill = True, element = 'step', palette = 'tab10', alpha=0.5)
plt.title("Wages by Gender")
plt.xlabel("Wages")
```

```python
# still not too clear, so let's use a boxplot to ensure:
sns.boxplot(data = the_data, x = 'gender', y = 'wage')
#sns.boxplot(data = the_data, x = 'age', y = 'wage')
```

```python
# however, this can be due to age or education levels, so let's take a look at those: 
sns.lmplot(data = the_data, x = "education", y = "wage", hue = "gender",
          scatter_kws = {"s": 5})
plt.xlabel("Education")
plt.ylabel("Wages")
plt.title("Wages vs. Education by Gender")
```

```python
# this could potentially be due to education, so lets look at that: 
sns.lmplot(data = the_data, x = "age", y = "wage", hue = "gender", 
          scatter_kws = {"s": 5}, order = 2)
plt.xlabel("Age")
plt.ylabel('Wages')
plt.title("Wages vs. Age by Gender")
```

```python
sns.lmplot(data = the_data, x = "age", y = "wage", hue = "gender", 
          scatter_kws = {"s": 5})
plt.xlabel("Age")
plt.ylabel('Wages')
plt.title("Wages vs. Age by Gender")
```

```python
# creating a KDE plot to see the data distributions of both genders
sns.displot(the_data, kind = 'kde', x = 'wage', hue = 'gender', fill = "True")
```

```python
sns.pairplot(the_data, hue = 'gender', palette = 'magma')
```

```python
g =sns.jointplot(x = 'age', y = 'wage', data = the_data, 
                 kind = 'kde', fill = True, hue = 'gender', 
                 alpha = 0.8, 
                palette = 'tab10')
g.ax_joint.set_xlabel('Age')
g.ax_joint.set_ylabel('Wages')
g.fig.suptitle("Joint Plot of Wages by Age and Gender")
g.fig.subplots_adjust(top=0.95) # Reduce plot to make room 
```

```python
g = sns.jointplot(x = 'education', y = 'wage', data = the_data, kind ='kde', fill = True, hue = 'gender', alpha = 0.8, 
              palette = 'tab10')
g.ax_joint.set_xlabel('Education')
g.ax_joint.set_ylabel('Wages')
g.fig.suptitle("Joint Plot of Wages by Education and Gender")
g.fig.subplots_adjust(top=0.95) # Reduce plot to make room 
```
